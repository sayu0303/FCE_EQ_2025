```python
import pandas as pd

```

##**Exemplo de aplicação para uma planilha de página única**

```python
#Análise de dados para uma planilha de página única:

#URL da planilha no GitHub
url = 'https://github.com/sayu0303/FCE_EQ_2025/raw/refs/heads/main/DADOS%20CANNA%20INDICA.xlsx'

#Leitura da planilha
df_dados = pd.read_excel(url, header = 1)

#Valores de referência
s_eb = 1.1389   #Sólidos totais - efluente bruto (mg/mL)
t_eb = 304      #Turbidez - efleunte bruto (NTU)


##Processa os dados planilha e mostra o resultado em uma nova planilha com as operações matemáticas desejadas
df_dados['Remoção de Sólidos (%)'] = 100 - ((df_dados['SÓLIDOS TOTAIS (mg/mL)']/s_eb)*100)
df_dados['Remoção de turbidez(%)'] = 100 - ((df_dados['TURBIDEZ (NTU)']/t_eb)*100)

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

df_dados

```

#**Aplicação para uma planilha de múltiplas páginas**

```python
#Análise de dados para planilha de múltiplas páginas

#URL da planilha
url = 'https://github.com/sayu0303/FCE_EQ_2025/raw/refs/heads/main/DADOS%20IC.xlsx'

# Lendo todas as abas

df_dados = pd.read_excel(url, sheet_name=None, header=1)

# Valores de referência
s_eb = 1.1389     #Sólidos totais - efluente bruto (mg/mL)
t_eb = 304        #Turbidez - efleunte bruto (NTU


pd.set_option('display.expand_frame_repr', False)  # não quebra linhas

#Processa cada página da planilha e mostra o resultado em uma nova planilha com as operações matemáticas desejadas
for tanque, df_dados in df_dados.items():
    df_dados['Remoção de Sólidos (%)'] = 100 - ((df_dados['SÓLIDOS TOTAIS (mg/mL)'] / s_eb) * 100)
    df_dados['Remoção de Turbidez (%)'] = 100 - ((df_dados['TURBIDEZ (NTU)'] / t_eb) * 100)

    print(f"--- {tanque} ---")
    print(df_dados, "\n")

```
