###a)
A linguagem compilada converte todo o código de uma vez para gerar um arquivo executável, através de um compilador. O código compilado tende a ser rápido, pois já está traduzido para o formato da máquina. Porém, os erros só são identificados no fim do processo de compilação, tornando a depuração lenta. Já a linguagem interpretada traduz o código linha por linha durante a execução do programa, sem gerar um arquivo executável. Essa execução é mais lenta, pois a tradução ocorre em tempo real enquanto o código está sendo executado.

###b)
A programação imperativa foca no passo a passo da execução, definindo a sequência de comandos para a máquina executar. Apesar de ser simples e direta, esse formato gera códigos longos, repetitivos e engessados, comprometendo a organização e eficiência do código. Agora, a programação procedural é um subtipo da imperativa, focando em funções ou procedimentos para dividir o código em blocos reutilizáveis. Essas características permitem que o código seja mais organizado e eficiente na manutenção, isolando problemas e reduzindo repetições.

```python
# Importando bibliotecas

import numpy as np

# Definir a função geradora das pressões

def GERADORpressao ():

  """
  Gerador de pressões aleatórias dentro do
  intervalo de 1.0 bar e 6.0 bar.
  """

  # Variável que irá gerar pressões diversar dentro do intervalo imposto

  pressaoAleatoria = np.random.uniform(1.0,6.0)
  return pressaoAleatoria
```

```python
# Definir a função verificadora das pressões

def VERIFICADORpressao(pressaoAleatoria):

  """
  Verificador se a pressão se encontra está FORA do padrão
  de segunrança, que é entre 2.0 e 5.0 bar. Caracterizando
  como ALERTA.
  """

  return pressaoAleatoria < 2.0 or pressaoAleatoria > 5.0
```

```python
def sistemaMONITOR ():
# Definir o contador de alarmes e leituras
  NUMleituras = 1
  alarmes = 0

# Definir a função verificadora das pressões

  # Para fazer as leituras

  for NUMleituras in range (10):
    NUMleituras = NUMleituras + 1
    pressaoAleatoria = GERADORpressao()

    if VERIFICADORpressao(pressaoAleatoria):
      alarmes = alarmes + 1

    if alarmes < 3:
      print(f"Leitura {NUMleituras}: Pressão: {pressaoAleatoria:.1f} bar, Alarmes Ativados: {alarmes}/3, Sistema: Operando")

    else:
      print(f"Leitura {NUMleituras}: Pressão: {pressaoAleatoria:.1f} bar, Alarmes Ativados: {alarmes}/3, Sistema: Interrompido")

      break
sistemaMONITOR()
```
